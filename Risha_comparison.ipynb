### Exploring differences

```python
diff_o_e = {}
    
for col in different_cols:
    diff_o_e["diff{0}".format(col)] = pd.concat([OFSP_original.iloc[:,col], OFSP_English.iloc[:,col]], axis = 1)
# this creates a dictionary of set differences by column index 
```

```python
diff_o_e.keys()
```

```python
for key in diff_o_e.keys():
    print(key, 'length', len(diff_o_e[key]))
    
# Each key is a df with 2 columns, one containing column data from OFSP_original and the other from OFSP_English and all rows
```

```python
globals().update(diff_o_e)
for key in diff_o_e.keys():
    x = globals()[key]
    diff_o_e[key] = x[x.iloc[:,0] != x.iloc[:,1]].dropna()
    
# This iterates through the keys (dfs) and drops duplicates where the value at OFSP_original is the same as OFSP English
# It also drops NAs
#Only rows where the 2 dfs have different values are retained
```

```python
for key in diff_o_e.keys():
    print(key, 'length', len(diff_o_e[key]))
 # lengths of the dfs
```
